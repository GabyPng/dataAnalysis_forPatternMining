* **Creadores:** Marko Bohanec y Blaz Zupan.

## 2. Estructura del Modelo

El dataset se basa en una jerarquía de decisión. Aunque el archivo final solo relaciona los atributos de entrada con la clase final, el modelo original sigue esta lógica:

| Nivel de Entrada | Concepto Intermedio | Objetivo Final (Target) |
| --- | --- | --- |
| `buying` (Precio de compra) | **PRICE** (Precio general) | **CAR** (Aceptabilidad) |
| `maint` (Mantenimiento) |  |  |
| `doors` (Puertas) | **COMFORT** (Confort) |  |
| `persons` (Capacidad) |  |  |
| `lug_boot` (Maletero) |  |  |
| `safety` (Seguridad) | **TECH** (Características Técnicas) |  |

---
# Cardinalidad y Dominio

## 3. Atributos (Variables de Entrada)


| Atributo  | Descripción                     | Dominio                                   | Cardinalidad |
|-----------|----------------------------------|--------------------------------------------|--------------|
| buying    | Precio de compra                 | {muy alto, alto, medio, bajo}              | 4            |
| maint     | Precio de mantenimiento          | {muy alto, alto, medio, bajo}              | 4            |
| doors     | Número de puertas                | {dos, tres, cuatro, mas}                   | 4            |
| persons   | Capacidad de personas            | {dos, cuatro, mas}                         | 3            |
| lug_boot  | Tamaño del maletero              | {pequeño, medio, grande}                   | 3            |
| safety    | Seguridad estimada del auto      | {bajo, medio, alto}                        | 3            |
|-----------|----------------------------------|--------------------------------------------|--------------|
| **Total combinaciones posibles** |  |  | **1728 (4×4×4×3×3×3)** |

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
from ucimlrepo import fetch_ucirepo 
  


# fetch dataset 
car_evaluation = fetch_ucirepo(id=19) 
  
# data (as pandas dataframes) 
X = car_evaluation.data.features 
y = car_evaluation.data.targets 
  
# metadata 
print(car_evaluation.metadata) 
  
# variable information 
print(car_evaluation.variables) 

df = pd.concat([car_evaluation.data.features, car_evaluation.data.targets], axis=1)

df.head()
```

```python

print("Tipo de datos por columna:")
df.info()


print("-"*60)
print("Numero de valores unicos por columna:")
print(df.nunique())


```

```python
print("Numero de instancias por clase | porcentaje:")
print(df['class'].value_counts())
print(df['class'].value_counts(normalize=True))
```

```python
print("-"*60)
print("Numero de Valores nulos por columna:")
print(df.isnull().sum())

print("-"*60)
print("Total de filas:", len(df))

print("-"*60)
df.describe(include='all')
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Reemplazar los valores de la columna doors a español
df['doors'] = df['doors'].replace({'2': 'dos', '3': 'tres', '4': 'cuatro', '5more': 'mas'})

# Reemplazar los valores de la columna persons a español
df['persons'] = df['persons'].replace({'2': 'dos', '4': 'cuatro', 'more': 'mas'})

# Traducir los valores de la columna maint a español
df['maint'] = df['maint'].replace({'vhigh': 'muy alto', 'high': 'alto', 'med': 'medio', 'low': 'bajo'})

# Traducir los valores de la columna buying a español
df['buying'] = df['buying'].replace({'vhigh': 'muy alto', 'high': 'alto', 'med': 'medio', 'low': 'bajo'})

# Traducir los valores de la columna lug_boot a español
df['lug_boot'] = df['lug_boot'].replace({'small': 'pequeño', 'med': 'medio', 'big': 'grande'})

# Traducir los valores de la columna safety a español
df['safety'] = df['safety'].replace({'low': 'bajo', 'med': 'medio', 'high': 'alto'})

# Traducir los valores de la columna class a español
df['class'] = df['class'].replace({'unacc': 'inaceptable', 'acc': 'aceptable', 'good': 'bueno', 'vgood': 'muy bueno'})

df.head()
```

```python
# Exportar en la carpeta carEvaluation los archivos csv
df.head(10).to_csv('carEvaluation/car_evaluation_10.csv', index=False, encoding='utf-8-sig')
df.head(100).to_csv('carEvaluation/car_evaluation_100.csv', index=False, encoding='utf-8-sig')
df.head(1000).to_csv('carEvaluation/car_evaluation_1000.csv', index=False, encoding='utf-8-sig')
df.to_csv('carEvaluation/car_evaluation_all.csv', index=False, encoding='utf-8-sig')
```

```python
with open("carEvaluation/car_evaluation.txt", "w", encoding='utf-8') as f:
    f.write(df.to_string(index=False))

print("¡Archivo de texto tipo tabla generado!")
```
